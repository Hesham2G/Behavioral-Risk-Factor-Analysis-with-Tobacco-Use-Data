```python
'''The 'Gender' column is divided into exactly three parts. Let's assume the survey was split equally between genders. 
This means we'll create masks for half of the 'Overall' values, and assign them the values 'Male' and 'Female'.'''


mask = np.random.choice([True, False], size=len(data))


data.loc[mask, 'Gender'] = 'Male'


data.loc[~mask, 'Gender'] = 'Female'
```

```python
categorical_features = ['TopicDesc','Race', 'MeasureDesc', 'Education', 'Gender']

data = pd.get_dummies(data, columns=categorical_features, drop_first=True)
```

```python
unique_age_values = data['Age'].unique()
print(unique_age_values)
```

```python
age_groups = ['All Ages', 'Age 25 and Older', 'Age 20 and Older', '18 to 24 Years',
              '18 to 44 Years', '45 to 64 Years', '25 to 44 Years', '65 Years and Older']

# Apply one-hot encoding to the 'age_group' column
encoded_age_groups = pd.get_dummies(data['Age'], prefix='age_groups')

# Concatenate the one-hot encoded columns with the original DataFrame
data = pd.concat([data, encoded_age_groups], axis=1)

# Drop the original 'age_group' column
data.drop(columns=['Age'], inplace=True)
```

```python
X = data.drop(['High_Confidence_Limit', 'LocationDesc', 'TopicType', 'Data_Value_Type'], axis=1)
```

```python
y = data['High_Confidence_Limit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=101)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```
